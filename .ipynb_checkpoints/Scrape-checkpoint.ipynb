```python
def capture_info(key_name, location_type, json_dict):
    places = jd.decode(json_dict[0]['DB_HOUSING_HELP_V2'])[key_name]['data']
    addresses = [place['address'] for place in places]
    postal_codes = [' '.join(address.split(' ')[-2:]) for address in addresses]
    provinces = [address.split(' ')[-3] for address in addresses]
    cities = [address.split(' ')[-5] for address in addresses]
    address_1s = [' '.join(address.split(' ')[:-5]) for address in addresses]
    clean_addresses = []
    for i, x in enumerate(places):
        clean_addresses.append({
            'address_1': address_1s[i],
            'city': cities[i],
            'postal_code': postal_codes[i],
            'province': provinces[i],
            'location_type': location_type,
            'eligibility': places[i]['eligibilityNotes'],
            'email': places[i]['email'],
            'phone_number': ''.join(re.findall(r'\d+', places[i]['phone'])),
            'name': places[i]['programName'],
            'url': places[i]['weburl'],
            'latitude': places[i]['latitude'],
            'longitude': places[i]['longitude']
        })
    return clean_addresses
```

```python
capture_info('ootc', 'Out of the Cold', json)
```
